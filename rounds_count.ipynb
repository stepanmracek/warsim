```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def read_rounds(path):
    with open(path, "r") as f:
        for line in f:
            yield int(line)

plt.title("Histogram of number of rounds of finished games")
plt.hist(list(read_rounds("rounds_no_shuffle.txt")), range=(0, 400), bins=200, label="w/o shuffle", alpha=0.5, density=True)
plt.hist(list(read_rounds("rounds.txt")), range=(0, 400), bins=200, label="with shuffle", alpha=0.5, density=True)
plt.legend()
plt.show()
```

```python
np.mean(list(read_rounds("rounds_no_shuffle.txt"))), np.mean(list(read_rounds("rounds.txt")))
```

```python
np.median(list(read_rounds("rounds_no_shuffle.txt"))), np.median(list(read_rounds("rounds.txt")))
```
